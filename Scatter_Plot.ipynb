Task 1: Exploring Relationships with Scatter Plots
1. Scatter Plot for Age vs. Job Satisfaction
Visualize the relationship between respondents' age (Age) and job satisfaction (JobSatPoints_6). Use this plot to identify any patterns or trends.

2. Scatter Plot for Compensation vs. Job Satisfaction
Explore the relationship between yearly compensation (ConvertedCompYearly) and job satisfaction (JobSatPoints_6) using a scatter plot.

```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/QR9YeprUYhOoLafzlLspAw/survey-results-public.sqlite
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect('survey-results-public.sqlite')

# Charger les données
QUERY = "SELECT * FROM main"
df = pd.read_sql_query(QUERY, conn)

### 1. Scatter Plot : Âge vs. Satisfaction au travail ###
if 'Age' in df.columns and 'JobSatPoints_6' in df.columns:
    df_filtered = df[['Age', 'JobSatPoints_6']].dropna()

    # Convertir Age en numérique
    #df_filtered['Age'] = pd.to_numeric(df_filtered['Age'], errors='coerce')

    # Tracer le scatter plot
    plt.figure(figsize=(10, 6))
 #=>
    plt.scatter(df_filtered['Age'], df_filtered['JobSatPoints_6'], alpha=0.5, color='blue')

    plt.title('Scatter Plot: Age vs. Job Satisfaction')
    plt.xlabel('Age')
    plt.ylabel('Job Satisfaction (JobSatPoints_6)')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
else:
    print("Les colonnes nécessaires ('Age' ou 'JobSatPoints_6') ne sont pas présentes dans la base.")

### 2. Scatter Plot : Compensation vs. Satisfaction au travail ###
if 'ConvertedCompYearly' in df.columns and 'JobSatPoints_6' in df.columns:
    df_filtered = df[['ConvertedCompYearly', 'JobSatPoints_6']].dropna()

    # Tracer le scatter plot
    plt.figure(figsize=(10, 6))
#=>
   plt.scatter(df_filtered['ConvertedCompYearly'], df_filtered['JobSatPoints_6'], alpha=0.5, color='green')

    plt.title('Scatter Plot: Compensation vs. Job Satisfaction')
    plt.xlabel('Yearly Compensation')
    plt.ylabel('Job Satisfaction (JobSatPoints_6)')
    plt.xscale('log')  # Compensation peut varier énormément, échelle logarithmique recommandée
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
else:
    print("Les colonnes nécessaires ('ConvertedCompYearly' ou 'JobSatPoints_6') ne sont pas présentes dans la base.")

# Fermer la connexion
conn.close()
```

##Task 2: Enhancing Scatter Plots
1. Scatter Plot with Trend Line for Age vs. Job Satisfaction
Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.

2. Scatter Plot for Age vs. Work Experience
Visualize the relationship between Age (Age) and Work Experience (YearsCodePro) using a scatter plot.

```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/QR9YeprUYhOoLafzlLspAw/survey-results-public.sqlite
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Connect to the SQLite database
conn = sqlite3.connect('survey-results-public.sqlite')

# Charger les données
QUERY = "SELECT * FROM main"
df = pd.read_sql_query(QUERY, conn)

### 1. Scatter Plot avec régression : Âge vs. Satisfaction au travail ###
if 'Age' in df.columns and 'JobSatPoints_6' in df.columns:
    df_filtered = df[['Age', 'JobSatPoints_6']].dropna()
    df_filtered['Age'] = pd.to_numeric(df_filtered['Age'], errors='coerce')

    # Tracer le scatter plot avec une régression
    plt.figure(figsize=(10, 6))
#=>
    sns.regplot(x=df_filtered['Age'], y=df_filtered['JobSatPoints_6'], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})

    plt.title('Scatter Plot with Trend Line: Age vs. Job Satisfaction')
    plt.xlabel('Age')
    plt.ylabel('Job Satisfaction (JobSatPoints_6)')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
else:
    print("Les colonnes nécessaires ('Age' ou 'JobSatPoints_6') ne sont pas présentes dans la base.")

### 2. Scatter Plot : Age vs. Work Experience ###
if 'Age' in df.columns and 'YearsCodePro' in df.columns:
    df_filtered = df[['Age', 'YearsCodePro']].dropna()
#   df_filtered['Age'] = pd.to_numeric(df_filtered['Age'], errors='coerce')
    df_filtered['YearsCodePro'] = pd.to_numeric(df_filtered['YearsCodePro'], errors='coerce')

    # Tracer le scatter plot
    plt.figure(figsize=(10, 6))
 #=>
    plt.scatter(df_filtered['Age'], df_filtered['YearsCodePro'], alpha=0.5, color='green')

    plt.title('Scatter Plot: Age vs. Work Experience')
    plt.xlabel('Age')
    plt.ylabel('Years of Professional Coding Experience')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
else:
    print("Les colonnes nécessaires ('Age' ou 'YearsCodePro') ne sont pas présentes dans la base.")

# Fermer la connexion
conn.close()
```

##Task 3: Combining Scatter Plots with Additional Features
1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size
Create a bubble plot to explore the relationship between yearly compensation (ConvertedCompYearly) and job satisfaction (JobSatPoints_6), with bubble size representing age.

 2. Scatter Plot for Popular Programming Languages by Job Satisfaction
Visualize the popularity of programming languages (LanguageHaveWorkedWith) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.

```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/QR9YeprUYhOoLafzlLspAw/survey-results-public.sqlite
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Connect to the SQLite database
conn = sqlite3.connect('survey-results-public.sqlite')

# Charger les données
QUERY = "SELECT * FROM main"
df = pd.read_sql_query(QUERY, conn)

### 1. Bubble Plot : Compensation vs. Job Satisfaction avec Age ###
if {'ConvertedCompYearly', 'JobSatPoints_6', 'Age'}.issubset(df.columns):
    df_filtered = df[['ConvertedCompYearly', 'JobSatPoints_6', 'Age']].dropna()

    # Normalisation de la taille des bulles (pour éviter des bulles trop grandes)
    df_filtered['Age'] = pd.to_numeric(df_filtered['Age'], errors='coerce')
    bubble_size = (df_filtered['Age'] - df_filtered['Age'].min()) / (df_filtered['Age'].max() - df_filtered['Age'].min()) * 500

    # Tracer le bubble plot
    plt.figure(figsize=(10, 6))
    plt.scatter(df_filtered['ConvertedCompYearly'], df_filtered['JobSatPoints_6'], s=bubble_size, alpha=0.5, color='blue')

    plt.title('Bubble Plot: Compensation vs. Job Satisfaction (Bubble Size = Age)')
    plt.xlabel('Yearly Compensation')
    plt.ylabel('Job Satisfaction (JobSatPoints_6)')
    plt.xscale('log')  # Échelle logarithmique pour la compensation
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
else:
    print("Les colonnes nécessaires ('ConvertedCompYearly', 'JobSatPoints_6' ou 'Age') sont absentes.")

### 2. Scatter Plot : Langages populaires vs. Satisfaction au travail ###
if {'LanguageHaveWorkedWith', 'JobSatPoints_6'}.issubset(df.columns):
    df_filtered = df[['LanguageHaveWorkedWith', 'JobSatPoints_6']].dropna()

    # Extraction des langages les plus populaires (top 10)
    lang_counts = df_filtered['LanguageHaveWorkedWith'].str.split(';').explode().value_counts()
    top_languages = lang_counts.head(10).index

    # Filtrer les données pour les langages les plus populaires
    df_filtered = df_filtered[df_filtered['LanguageHaveWorkedWith'].apply(lambda x: any(lang in x for lang in top_languages))]

    # Transformation pour affichage (chaque langage devient une ligne)
    df_exploded = df_filtered.assign(Language=df_filtered['LanguageHaveWorkedWith'].str.split(';')).explode('Language')

    # Garder uniquement les langages du top 10
    df_exploded = df_exploded[df_exploded['Language'].isin(top_languages)]

    # Tracer le scatter plot
    plt.figure(figsize=(12, 6))
    sns.stripplot(x='Language', y='JobSatPoints_6', data=df_exploded, jitter=True, alpha=0.5)

    plt.title('Scatter Plot: Popular Programming Languages by Job Satisfaction')
    plt.xlabel('Programming Language')
    plt.ylabel('Job Satisfaction (JobSatPoints_6)')
    plt.xticks(rotation=45)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
else:
    print("Les colonnes nécessaires ('LanguageHaveWorkedWith' ou 'JobSatPoints_6') sont absentes.")

# Fermer la connexion
conn.close()
```

##Task 4: Scatter Plot Comparisons Across Groups
1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type
Visualize the relationship between yearly compensation (ConvertedCompYearly) and job satisfaction (JobSatPoints_6), categorized by employment type (Employment). Use color coding or markers to differentiate between employment types.

2. Scatter Plot for Work Experience vs. Age Group by Country
Compare work experience (YearsCodePro) across different age groups (Age) and countries (Country). Use colors to represent different countries and markers for age groups.

```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/QR9YeprUYhOoLafzlLspAw/survey-results-public.sqlite
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Connect to the SQLite database
conn = sqlite3.connect('survey-results-public.sqlite')

# Charger les données
QUERY = "SELECT * FROM main"
df = pd.read_sql_query(QUERY, conn)

### 1. Scatter Plot : Compensation vs. Job Satisfaction par Type d'Emploi ###
if {'ConvertedCompYearly', 'JobSatPoints_6', 'Employment'}.issubset(df.columns):
    df_filtered = df[['ConvertedCompYearly', 'JobSatPoints_6', 'Employment']].dropna()

    # Tracer le scatter plot
    plt.figure(figsize=(12, 6))
    sns.scatterplot(x='ConvertedCompYearly', y='JobSatPoints_6', hue='Employment', data=df_filtered, alpha=0.6)

    plt.title('Scatter Plot: Compensation vs. Job Satisfaction by Employment Type')
    plt.xlabel('Yearly Compensation')
    plt.ylabel('Job Satisfaction (JobSatPoints_6)')
    plt.xscale('log')  # Compensation très variable, échelle log recommandée
    plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
else:
    print("Les colonnes nécessaires ('ConvertedCompYearly', 'JobSatPoints_6' ou 'Employment') sont absentes.")

### 2. Scatter Plot : Expérience vs. Groupe d'Âge par Pays ###
if {'YearsCodePro', 'Age', 'Country'}.issubset(df.columns):
    df_filtered = df[['YearsCodePro', 'Age', 'Country']].dropna()

    # Convertir en numérique
    df_filtered['YearsCodePro'] = pd.to_numeric(df_filtered['YearsCodePro'], errors='coerce')
    df_filtered['Age'] = pd.to_numeric(df_filtered['Age'], errors='coerce')

    # Définir des groupes d'âge
    bins = [18, 25, 35, 45, 55, 65, 100]
    labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
    df_filtered['AgeGroup'] = pd.cut(df_filtered['Age'], bins=bins, labels=labels, right=False)

    # Tracer le scatter plot
    plt.figure(figsize=(12, 6))
    sns.scatterplot(x='YearsCodePro', y='AgeGroup', hue='Country', data=df_filtered, alpha=0.6)

    plt.title('Scatter Plot: Work Experience vs. Age Group by Country')
    plt.xlabel('Years of Professional Coding Experience')
    plt.ylabel('Age Group')
    plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
else:
    print("Les colonnes nécessaires ('YearsCodePro', 'Age' ou 'Country') sont absentes.")

# Fermer la connexion
conn.close()
```
