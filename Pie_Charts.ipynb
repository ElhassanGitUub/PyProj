###Task 1: Visualizing Data Composition with Pie Charts

 1.1 The DevType column lists the developer types for respondents.

 1.2 We’ll examine the distribution by showing the top 5 developer roles in a pie chart.

 1.3 Create a pie chart for the operating systems used by respondents for professional use
The OpSysProfessional use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)
# 1.1 The DevType column lists the developer types for respondents.

# Split multiple developer roles (some respondents have multiple roles separated by ";")
df["DevType"] = df["DevType"].str.split(";")

# Expand the DevType column into multiple rows
df_exploded = df.explode("DevType")

# Count the occurrences of each developer type
devtype_counts = df_exploded["DevType"].value_counts().head(5)  # Top 5 roles

# Plot Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(devtype_counts, labels=devtype_counts.index, autopct="%1.1f%%", colors=["#ff9999","#66b3ff","#99ff99","#ffcc99","#c2c2f0"])
plt.title("Top 5 Developer Roles")
plt.show()

# 1.2 We’ll examine the distribution by showing the top 5 developer roles in a pie chart.
# Count occurrences of each OS
opsys_counts = df["OpSysProfessional use"].value_counts().head(5)  # Top 5 OS

# Plot Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(opsys_counts, labels=opsys_counts.index, autopct="%1.1f%%", colors=["#ff9999","#66b3ff","#99ff99","#ffcc99","#c2c2f0"])
plt.title("Operating Systems Used for Professional Work")
plt.show()

```

###Task 3: Analyzing and Interpreting Composition
In this task, you will create additional pie charts to analyze specific aspects of the survey data. Use `pandas` and `matplotlib` to complete each task and interpret the findings.


- 3.1 Pie Chart of `Respondents` Most Admired Programming Languages
The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.

- 3.2 Pie Chart of Tools Used for AI Development
Using the AIToolCurrently Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.


- 3.3 Pie Chart for Preferred Web Frameworks
The WebframeWantToWorkWith column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.

- 3.4 Pie Chart for Most Desired Embedded Technologies
Using the EmbeddedWantToWorkWith column, create a pie chart to show the top 5 most desired embedded technologies that respondents wish to work with.

```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/QR9YeprUYhOoLafzlLspAw/survey-results-public.sqlite
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connexion à la base SQLite
conn = sqlite3.connect('survey-results-public.sqlite')

# Charger les données
QUERY = "SELECT * FROM main"
df = pd.read_sql_query(QUERY, conn)

# Fonction pour générer un Pie Chart
def plot_pie_chart(column_name, title):
    if column_name in df.columns:
        df_filtered = df[[column_name]].dropna()

        # Séparer les valeurs multiples
        df_exploded = df_filtered.assign(Category=df_filtered[column_name].str.split(';')).explode('Category')

        # Compter les occurrences
        category_counts = df_exploded['Category'].value_counts().nlargest(5)  # Top 5

        # Tracer le pie chart
        plt.figure(figsize=(8, 6))
        category_counts.plot.pie(autopct='%1.1f%%', cmap='Set3', startangle=140)
        plt.title(title)
        plt.ylabel('')  # Supprimer le label
        plt.show()
    else:
        print(f"Colonne '{column_name}' non trouvée dans les données.")

### 1. Langages de programmation les plus admirés ###
plot_pie_chart('LanguageAdmired', 'Most Admired Programming Languages')

### 2. Outils IA les plus utilisés ###
plot_pie_chart('AIToolCurrentlyUsing', 'Top AI Development Tools')

### 3. Frameworks web les plus souhaités ###
plot_pie_chart('WebframeWantToWorkWith', 'Most Desired Web Frameworks')

### 4. Technologies embarquées les plus désirées ###
plot_pie_chart('EmbeddedWantToWorkWith', 'Most Desired Embedded Technologies')

# Fermer la connexion
conn.close()
```
